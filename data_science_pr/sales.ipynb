```python
import pandas as pd
import numpy as np
import chardet
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
with open('/content/drive/MyDrive/IMDb.csv', 'rb') as f:
    encoding = chardet.detect(f.read())['encoding']
```

```python
data=pd.read_csv('/content/drive/MyDrive/IMDb.csv',encoding=encoding)
data
```

```python
data.dropna(inplace=True)
data['Year'] = pd.to_numeric(data['Year'].str.extract('(\d+)')[0])
data['Votes'] = data['Votes'].str.replace(',', '').astype(int)
data
```

```python
from scipy import stats
z_scores = np.abs(stats.zscore(data[['Rating', 'Votes']]))
threshold = 3
outliers = np.where(z_scores > threshold)
outliers[0].shape
```

```python
#One potential bias related to directors or actors in the dataset could be the unequal representation of certain directors or actors compared to others. This bias could occur if certain directors or actors have a significantly higher number of movies in the dataset compared to others. This could skew any analysis or modeling towards the works of those directors or actors who are more prevalent in the dataset.
```

```python
from sklearn.utils import resample

director_distribution = data['Director'].value_counts()
balanced_data = pd.DataFrame()
for director in director_distribution.index:
    director_data = data[data['Director'] == director]
    if len(director_data) > 5:
        director_data_downsampled = resample(director_data, replace=False, n_samples=5, random_state=42)
    else:
        director_data_downsampled = director_data
    balanced_data = pd.concat([balanced_data, director_data_downsampled])
```

```python
distribution_balanced = balanced_data['Director'].value_counts()
distribution_balanced
```

```python
split_genres = balanced_data['Genre'].str.split(', ')
genre_columns = split_genres.apply(lambda x: pd.Series([1] * len(x), index=x)).fillna(0)
genre_counts = genre_columns.sum().astype(int)
print(genre_counts)
```

```python
genre_encoded = pd.get_dummies(balanced_data['Genre'].str.split(', ', expand=True).stack(), prefix='Genre').sum(level=0)
balanced_data = pd.concat([balanced_data, genre_encoded], axis=1)
balanced_data.drop('Genre', axis=1, inplace=True)
balanced_data
```

```python
from sklearn.preprocessing import LabelEncoder

director_encoder = LabelEncoder()
balanced_data['Director_Encoded'] = director_encoder.fit_transform(balanced_data['Director'].astype(str))

actors = balanced_data[['Actor 1', 'Actor 2', 'Actor 3']].values.ravel()
actor_encoder = LabelEncoder()
actor_encoder.fit(actors)
for i in range(1, 4):
    balanced_data[f'Actor_{i}_Encoded'] = actor_encoder.transform(balanced_data[f'Actor {i}'].fillna('').astype(str))

balanced_data.head()
```

```python
selected_features = ['Year', 'Duration', 'Votes','Director_Encoded','Actor_1_Encoded', 'Actor_2_Encoded', 'Actor_3_Encoded']+['Genre_Action',
 'Genre_Adventure',
 'Genre_Animation',
 'Genre_Biography',
 'Genre_Comedy',
 'Genre_Crime',
 'Genre_Documentary',
 'Genre_Drama',
 'Genre_Family',
 'Genre_Fantasy',
 'Genre_History',
 'Genre_Horror',
 'Genre_Music',
 'Genre_Musical',
 'Genre_Mystery',
 'Genre_News',
 'Genre_Romance',
 'Genre_Sci-Fi',
 'Genre_Sport',
 'Genre_Thriller',
 'Genre_War',
 'Genre_Western']
balanced_data['Rating'] = balanced_data['Rating'].round().astype(int)
balanced_data['Duration'] = balanced_data['Duration'].str.extract(r'(\d+)').astype(float)

X = balanced_data[selected_features]
Y = balanced_data.Rating
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
X_train
```
